```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
house_data = pd.read_csv("2018 House Popular Vote Tracker - Sheet1.csv")[2:]
house_data.head()
```

```python
all_data = pd.read_csv("us-midterms-18.csv", sep = ';')
all_data.head()
```

```python
senate_data = all_data[(all_data['Office']=='Senate')&(all_data['Election Type']=='general')]
print(senate_data.shape)
senate_data.head()
```

```python
gov_data = all_data[(all_data['Office']=='Governor')&(all_data['Election Type']=='general')]
print(gov_data.shape)
gov_data.head()
```

```python
mixed_national = dict()
for state in house_data['State'].unique():
    state_data = house_data[house_data['State']==state]
    uncontested_data = state_data[(state_data['Dem Votes']==0)|(state_data['GOP Votes']==0)|
                                  (state_data['Other Votes']>state_data['Dem Votes'])|
                                  (state_data['Other Votes']>state_data['GOP Votes'])]
    if len(uncontested_data) == 0:
        mixed_national[state] = "House"
    else:
        state_data = senate_data[senate_data['State']==state]
        uncontested_data = state_data[(state_data['Results Dem'].isnull())|(state_data['Results Rep'].isnull())]
        if len(state_data) > 0 and len(uncontested_data) == 0:
            mixed_national[state] = "Senate"
        else:
            state_data = gov_data[gov_data['State']==state]
            uncontested_data = state_data[(state_data['Results Dem'].isnull())|(state_data['Results Rep'].isnull())]
            if len(state_data) > 0 and len(uncontested_data) == 0:
                mixed_national[state] = "Governor"
            else:
                mixed_national[state] = "Other"
print(len(mixed_national))
mixed_national
```

```python
all_options = dict()
for state in house_data['State'].unique():
    state_data = house_data[house_data['State']==state]
    uncontested_data = state_data[(state_data['Dem Votes']==0)|(state_data['GOP Votes']==0)|
                                  (state_data['Other Votes']>state_data['Dem Votes'])|
                                  (state_data['Other Votes']>state_data['GOP Votes'])]
    if len(uncontested_data) == 0:
        all_options[state] = ["House"]
        
    state_data = senate_data[senate_data['State']==state]
    uncontested_data = state_data[(state_data['Results Dem'].isnull())|(state_data['Results Rep'].isnull())]
    if len(state_data) > 0 and len(uncontested_data) == 0:
        if state in all_options: all_options[state].append("Senate")
        else: all_options[state] = ["Senate"]

    state_data = gov_data[gov_data['State']==state]
    uncontested_data = state_data[(state_data['Results Dem'].isnull())|(state_data['Results Rep'].isnull())]
    if len(state_data) > 0 and len(uncontested_data) == 0:
        if state in all_options: all_options[state].append("Governor")
        else: all_options[state] = ["Governor"]
    if state not in all_options: all_options[state] = ["Other"]
print(len(all_options))
all_options
```

House Popular Vote

```python
dem_count = 0
gop_count = 0
other_count = 0
for state in mixed_national:
    dem_count += house_data[house_data['State']==state]['Dem Votes'].sum()
    gop_count += house_data[house_data['State']==state]['GOP Votes'].sum()
    other_count += house_data[house_data['State']==state]['Other Votes'].sum()
total = dem_count + gop_count + other_count
print(dem_count)
print(gop_count)
print(dem_count - gop_count)
house_dem_margin = dem_count / total - gop_count / total
house_dem_margin
```

Mixed National Popular Vote

```python
dem_count = 0
gop_count = 0
other_count = 0
for state in mixed_national:
    if mixed_national[state] == "House":
        dem_count += house_data[house_data['State']==state]['Dem Votes'].sum()
        gop_count += house_data[house_data['State']==state]['GOP Votes'].sum()
        other_count += house_data[house_data['State']==state]['Other Votes'].sum()
    elif mixed_national[state] == "Senate":
        dem_count += senate_data[(senate_data['State']==state)&
                                 (senate_data['Candidate partyname']=='Democratic')]['Candidate votes'].sum()
        gop_count += senate_data[(senate_data['State']==state)&
                                 (senate_data['Candidate partyname']=='Republican')]['Candidate votes'].sum()
        other_count += senate_data[(senate_data['State']==state)&
                                 (senate_data['Candidate partyname']!='Democratic')&
                                  (senate_data['Candidate partyname']!='Republican')]['Candidate votes'].sum()
    elif mixed_national[state] == "Governor":
        dem_count += gov_data[(gov_data['State']==state)&
                                 (gov_data['Candidate partyname']=='Democratic')]['Candidate votes'].sum()
        gop_count += gov_data[(gov_data['State']==state)&
                                 (gov_data['Candidate partyname']=='Republican')]['Candidate votes'].sum()
        other_count += gov_data[(gov_data['State']==state)&
                                 (gov_data['Candidate partyname']!='Democratic')&
                                  (gov_data['Candidate partyname']!='Republican')]['Candidate votes'].sum()       
    elif mixed_national[state] == "Other":
        if state == "North Carolina": # State Supreme Court
            # https://ballotpedia.org/North_Carolina_Supreme_Court_elections,_2018
            dem_count += 1811117
            gop_count += 1245901 + 598579
        elif state == "Louisiana": # Secretary of State Runoff
            # https://voterportal.sos.la.gov/static/2018-12-08/resultsRace/Statewide
            dem_count += 210085
            gop_count += 306568
    else:
        print(state)
total = dem_count + gop_count + other_count
print(dem_count)
print(gop_count)
print(dem_count - gop_count)
mixed_dem_margin = dem_count / total - gop_count / total
mixed_dem_margin
```

Simulations

```python
import random
random.seed(538)
```

```python
sim_data = {}
for state in all_options:
    sim_data[state] = []
    for option in all_options[state]:
        if option == "House":
            dem_count = house_data[house_data['State']==state]['Dem Votes'].sum()
            gop_count = house_data[house_data['State']==state]['GOP Votes'].sum()
            other_count += house_data[house_data['State']==state]['Other Votes'].sum()
        elif option == "Senate":
            dem_count = senate_data[(senate_data['State']==state)&
                                     (senate_data['Candidate partyname']=='Democratic')]['Candidate votes'].sum()
            gop_count = senate_data[(senate_data['State']==state)&
                                     (senate_data['Candidate partyname']=='Republican')]['Candidate votes'].sum()
            other_count = senate_data[(senate_data['State']==state)&
                                     (senate_data['Candidate partyname']!='Democratic')&
                                      (senate_data['Candidate partyname']!='Republican')]['Candidate votes'].sum()
        elif option == "Governor":
            dem_count = gov_data[(gov_data['State']==state)&
                                     (gov_data['Candidate partyname']=='Democratic')]['Candidate votes'].sum()
            gop_count = gov_data[(gov_data['State']==state)&
                                     (gov_data['Candidate partyname']=='Republican')]['Candidate votes'].sum()
            other_count = gov_data[(gov_data['State']==state)&
                                     (gov_data['Candidate partyname']!='Democratic')&
                                      (gov_data['Candidate partyname']!='Republican')]['Candidate votes'].sum()
        elif option == "Other":
            if state == "North Carolina": # State Supreme Court
                # https://ballotpedia.org/North_Carolina_Supreme_Court_elections,_2018
                dem_count = 1811117
                gop_count = 1245901 + 598579
                other_count = 0
            elif state == "Louisiana": # Secretary of State Runoff
                # https://voterportal.sos.la.gov/static/2018-12-08/resultsRace/Statewide
                dem_count = 210085
                gop_count = 306568
                other_count = 0
        sim_data[state].append([dem_count,gop_count,other_count])
sim_data
```

```python
dem_margins = []
n_sims = 10000
for trial in range(n_sims):
    dem_count = 0
    gop_count = 0
    other_count = 0
    for state in sim_data:
        ind = random.randint(0,len(sim_data[state])-1)
        option = sim_data[state][ind]
        dem_count += option[0]
        gop_count += option[1]
        other_count += option[2]
    total = dem_count + gop_count + other_count
    dem_margin = dem_count / total - gop_count / total
    dem_margins.append(dem_margin)
dem_margins = pd.Series(dem_margins)
print(dem_margins.mean())
print(dem_margins.std())

dem_margins.plot(kind='hist', color = "skyblue")
plt.axvline(x = house_dem_margin, color = 'navy', lw = 3)
plt.axvline(x = dem_margins.mean(), color = 'blue', lw = 3)
plt.xlabel("Democrat Victory Margin")
plt.title("Democrat Victory Margins")
plt.text(x = .061, y = 2000, s = "House Popular Vote", color = "navy")
plt.text(x = .02, y = 2000, s = "Mean of Simulations", color = "blue")
plt.text(x = .016, y = 1000, s = "Simulations", color = "dodgerblue")
plt.savefig("histogram.png")
plt.show()
```
